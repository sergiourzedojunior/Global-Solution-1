```python
import sqlite3
import pandas as pd
from pathlib import Path

# Caminho para o banco
DB_PATH = Path(r"C:\GitHub\Global Solution 1\db\gs2025.db")

# Abrir conex√£o
with sqlite3.connect(str(DB_PATH)) as con:
    
    print("üìå Tabelas existentes:")
    display(pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", con))
    
    print("\nüìÑ Estrutura da tabela 'igv_resultado_all':")
    estrutura = pd.read_sql("PRAGMA table_info(igv_resultado_all)", con)
    display(estrutura)

    print("\nüîç Amostra de dados IGV:")
    df_igv = pd.read_sql("SELECT * FROM igv_resultado_all LIMIT 50", con)
    display(df_igv)

    print("\nüìä Regi√µes com mais registros de IGV:")
    df_regioes = pd.read_sql("""
        SELECT regiao, COUNT(*) AS total
        FROM igv_resultado_all
        GROUP BY regiao
        ORDER BY total DESC
    """, con)
    display(df_regioes)
```

```python
df_top10 = pd.read_sql("""
    SELECT nome_udh, nome_mun, regiao, igv
    FROM igv_resultado_all
    ORDER BY igv DESC
    LIMIT 10
""", con)
display(df_top10)
```

```python
df_agrupado = pd.read_sql("""
    SELECT regiao, AVG(igv) AS media_igv, COUNT(*) AS total
    FROM igv_resultado_all
    GROUP BY regiao
    ORDER BY media_igv DESC
""", con)
display(df_agrupado)
```
