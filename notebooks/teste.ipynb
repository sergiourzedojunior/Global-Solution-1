```python
df_igv = pd.read_sql("SELECT * FROM igv_resultado ORDER BY igv DESC LIMIT 10", conn)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.barh(df_igv["nome_udh"], df_igv["igv"], color='tomato')
plt.xlabel("IGV (quanto maior, mais vulnerável)")
plt.title("Top 10 UDHs mais vulneráveis")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
from pathlib import Path

# Caminho base para a pasta 'dados'
base_dir = Path("../dados").resolve()  # ajuste o caminho se necessário

# Lista para armazenar os caminhos completos
arquivos = []

for file in base_dir.rglob("*"):
    if file.is_file():
        arquivos.append(str(file))

print(f"Total de arquivos encontrados: {len(arquivos)}")
for path in arquivos:
    print(path)
```
