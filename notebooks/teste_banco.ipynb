```python
import sqlite3
import pandas as pd
from pathlib import Path

# Caminho direto e seguro (modifique se necess√°rio)
DB_PATH = Path(r"C:\GitHub\Global Solution 1\db\gs2025.db")

print("üîç Verificando caminho...")
print("DB_PATH:", DB_PATH)
print("Existe?", DB_PATH.exists())

if DB_PATH.exists():
    with sqlite3.connect(str(DB_PATH)) as con:
        print("\nüìã Tabelas dispon√≠veis:")
        tabelas = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", con)
        display(tabelas)

        print("\nüìÑ Estrutura da tabela 'indicadores_udh_all':")
        estrutura = pd.read_sql("PRAGMA table_info(indicadores_udh_all)", con)
        display(estrutura)

        print("\nüîç Amostra de dados:")
        df_exemplo = pd.read_sql("SELECT * FROM indicadores_udh_all LIMIT 5", con)
        display(df_exemplo)

        print("\nüìä Contagem por regi√£o:")
        contagem = pd.read_sql("""
            SELECT regiao, COUNT(*) AS total
            FROM indicadores_udh_all
            GROUP BY regiao
            ORDER BY total DESC
        """, con)
        display(contagem)
else:
    print("‚ùå Caminho do banco inv√°lido ou inexistente.")
```
